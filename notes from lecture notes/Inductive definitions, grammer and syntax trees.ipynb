Inductive Definitions in Scala
</pre>

```scala
//inductive definition
sealed trait Number
/*
    sealed tells scala that everything else that inherits from the trait number will need to be in the same file
    trait in scala is like an abstract class, we cant created an instance of it but we can write classes that inherit from it
    */
case class Z() extends Number
/*
Z is a class that extends number and models number -> Z
we write case class instead of class due to convienience
*/
case class Succ(n: Number) extends Number
```

when stepping through
    first handles let binding, x = 20
    now environment = {x: 20.0 }
    now process function definition
    function def evalutates to closure, first is closure(z, z - x, env{x: 20.0 and Nil})
    then foo gets mapped to closure, env{foo:closure(z,...,...)}
    then binds x , adds env{x: 30.0}
    now processing function call, foo evaluates to closure, argument to function is 40, now function call gets executed
    then now environment becomes env{ z: 40.0 , x:20.0}

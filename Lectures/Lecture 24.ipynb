```scala
def fib(n: Int, b:Boolean = true): Int = {
    if(b){
        println(s"Fibbonacci($n) called")
    }
    if(n <= 2){
        1
    }else{
        fib(n-1,false) + fib(n-2,false)
    }
}
def fact(n:Int, b:Boolean = true): Int = {
    if(b){
        println(s"Factorial($n) called")
    }
    if(n <= 1){
        1
    }else{
        n*fact(n-1,false)
    }
}
//forcing lazy by 
def myfoo(x: => Int, y: => Int):Int = {
    lazy val x0 = x
    lazy val y0 = y
    if(scala.util.Random.nextInt >= 0){
        println(s"Value of myfoo is $x0")
        x0
    }else{
        val new_y = y
        println(s"Value of myfoo is $y0")
        y0
    }
}


//eager evaluation is evaluating all functions then passing to myfoo
//to get rid of this use lazy evaluation 
//myfoo(fib(10),fact(8))
def mybar: Int = {
    myfoo(fact(10),fib(8)) + myfoo(fact(10),fib(8))
}

val x = mybar
```

```scala
def While(b: => Boolean, body: => Unit){
    if(b){
        body;
        While(b,body)
    }
}
var x=0
While(x<10,{
    x = x+1;
    println(x)
})

//if we remove laziness, infinite loop b not changing 
def While2(b:Boolean, body: => Unit){
    if(b){
        body;
        While2(b,body)
    }
}

//var x1=0
//While2(x<10,{
  //  x1 = x+1;
    //println(x)
//})

def While3(b:=>Boolean, body:Unit){
    if(b){
        body;
        While2(b,body)
    }
}
var x3=0
While(x3<10,{
    x3 = x3+1;
    println(x3)
})
//loop still doesnt terminate, x will remain as one

```
